```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Activation, Dropout
from tensorflow.keras.layers import Conv1D, MaxPooling1D, GlobalMaxPooling1D
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from pandas import read_excel
```

```python
df = pd.read_csv('DatasetTweet.csv')
df.head()
```

```python
text = df["Tweet"].tolist()
print(text)
```

```python
y = df['sentimen']
y = to_categorical(y)
print(y)
```

```python
df['sentimen'].value_counts()
```

```python
token = Tokenizer()
token.fit_on_texts(text)
```

```python
token.index_word
```

```python
vocab = len(token.index_word)+1
vocab
```

```python
x = ['aku lagi galau nih']
token.texts_to_sequences(x)
```

```python
encode_text = token.texts_to_sequences(text)
```

```python
max_kata = 100
X = pad_sequences(encode_text, maxlen = max_kata, padding="post")
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=40, test_size=0.3, stratify=y)
```

```python
X_train = np.asarray(X_train)
X_test = np.asarray(X_test)
y_train = np.asarray(y_train)
y_test = np.asarray(y_test)
```

```python
vec_size = 300
model = Sequential()
model.add(Embedding(vocab, vec_size, input_length=max_kata))
model.add(Conv1D(64,8, activation="relu"))
model.add(MaxPooling1D(2))
model.add(Dropout(0.5))

model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(16, activation='relu'))
model.add(GlobalMaxPooling1D())
model.add(Dense(3, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test))  
```

```python
def get_encode(x):
  x = token.texts_to_sequences(x)
  x = pad_sequences(x, maxlen = max_kata, padding="post")
  return x
```

```python
x = ["ga jelas banget videonya, suaranya kecil, unsubscribelah"]
x = get_encode(x)
```

```python
predict_x=model.predict(x) 
classes_x=np.argmax(predict_x,axis=1)
classes_x
```
